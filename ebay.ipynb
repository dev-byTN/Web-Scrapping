```python
def getInfo():

    try :
        if soup.find('span', class_="ux-textspans ux-textspans--BOLD").get_text():
            title = soup.find('span', class_="ux-textspans ux-textspans--BOLD").get_text()
        else :
            title = 'NaN'
    except : 
        title = 'NaN'

    try :
        if soup.find('div', class_="x-price-primary").get_text():
            price = soup.find('div', class_="x-price-primary").get_text()
        else :
            price = 'NaN'
    except :
        price = 'NaN'

    try :
        if soup.find('span', class_="ux-icon-text__text").get_text():
            condition = soup.find('span', class_="ux-icon-text__text").get_text()
        else :
            condition = 'NaN'
    except :
         condition = 'NaN'

    try :
        if soup.find('div', class_="ux-layout-section__row").get_text():
            location = soup.find('div', class_="ux-layout-section__row").get_text()
        else :
            location = 'NaN'
    except :
        location = 'NaN'

    info = [title,price,condition,location]

    return info

    #print(f"{title} {price} {condition}")
    
```
