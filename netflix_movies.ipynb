```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
```

```python
dt = pd.read_csv('movies.csv')
dt
```

```python
dt.dtypes
```

```python
dt.isnull().sum()
```

```python
dt = dt.drop_duplicates()
```

```python
dt['GENRE'] = dt['GENRE'].str.replace('\n','') #on enleve les \n dans la colonne genre
dt
```

```python
dt['YEAR'] = dt['YEAR'].str.replace('[^0-9]','', regex = True) # '^[^%]'= % en debut
dt
```

```python
dt["YEAR"] = dt["YEAR"].apply(lambda year : str(year))
dt["YEAR"] = dt["YEAR"].apply(lambda year : (year[0:4] + ", " + year[4:8]) if len(year)>4 else year)
dt
```

```python
dt['ONE-LINE'] = dt['ONE-LINE'].str.replace('\n','')
dt['STARS'] = dt['STARS'].str.replace('\n','')
dt
```

```python
#on apperçoit que dans la colonne stars il y a les directors des films en question
#donc nous allons créer une nouvelle colonne directors et renseigner pour chaque films les directors en question
```

```python
dt.insert(5,'DIRECTOR', 0)
```

```python
dt[['DIRECTOR', 'STARS']] = dt['STARS'].str.split('Stars:', expand=True)
dt
```

```python
dt['DIRECTOR'] = dt['DIRECTOR'].str.replace('Director:', '')
dt['DIRECTOR'] = dt['DIRECTOR'].str.replace('|', '')
dt
```

```python
dt
```

```python
dt.reset_index(drop = True, inplace = True)
dt
copied = dt
```

```python
dt.drop_duplicates(subset=['MOVIES','YEAR','DIRECTOR','GENRE'])
dt
```

```python
dt = dt[dt['Gross'].notnull()]
dt.sort_values(by = 'Gross', ascending = True).head(10)
```

```python
dt['Gross'] = dt['Gross'].str.replace('$','')
dt['Gross'] = dt['Gross'].str.replace('M', '')
dt
```

```python
dt.dtypes
```

```python
Movie_Gross = dt.reset_index(drop = True, inplace = True)
Movie_Gross
```
