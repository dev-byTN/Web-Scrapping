```python
import pandas as pd
import numpy as np
```

```python
dt = pd.read_csv('flights.csv')
dt.head()
```

```python
#if flights in different year
dt.loc[dt['Year'] != 2013]
```

```python
#if flights in wrong dayofweek
dt.loc[dt['DayOfWeek'] > 7]
```

```python
#if flights in wrong dayofmonth
dt.loc[dt['DayofMonth'] > 31]
```

```python
#if flights in wrong month
dt.loc[dt['Month'] > 31]
```

```python
#si valeurs manquantes 
dt.isnull().sum()
```

```python
# on va regarder si wrong data dans les >15 minutes et 0 dans depdel15
condition = ((dt['DepDelay'] > 15) & (dt['DepDel15']<1))
dt.loc[condition]
```

```python
#on va regarder les valeurs manquantes dans depdel5
dt.loc[dt['DepDel15'].isnull()]
```

```python
#on a pu voir que les valeurs manquantes interviennent quand le depDelay = 0 ( pas de retard) donc depdel15 = 0
#on va d'abord confirmer ca
condition = ((dt['DepDelay'] == 0) & (dt['DepDel15']>0)) #nous affichons les valeurs avec un retard et avec un depdel15 NaN
dt.loc[condition]
```

```python
dt
```

```python
#nous pouvosn confirmer que les NaN interviennent quand ik n'y a pas de retard 
#nous allons donc modifier ces valeurs et les mettre à 0 comme il n'y a aucun retard de plus de 15mn
newdt = dt
newdt.loc[(newdt['DepDelay'] == 0), "DepDel15"] = 0
newdt.loc[dt['DepDelay'] == 0]
newdt
```

```python
newdt.isnull().sum() #we have change all the naN value
```

```python
#the arrDelay and arrDel15 are wrong we must fix it
newdt.loc[((newdt['ArrDelay'] < 15) & (newdt['ArrDel15'] == 1))]
```

```python
newdt.loc[((newdt['ArrDelay'] < 15) & (newdt['ArrDel15'] == 1)), "ArrDel15"] = 0 #on remplace par 0
newdt.loc[((newdt['ArrDelay'] > 15) & (newdt['ArrDel15'] != 1)), "ArrDel15"] = 1 #on remplace par 1
newdt.loc[((newdt['ArrDelay'] == 0) & (newdt['ArrDel15'] < 1))]
```

```python
#maintenant on regarde si il y a des vols annulés mais avec des heures d'arrivées et de départ
newdt.loc[((newdt['DepDelay'] !=0) & (newdt['ArrDelay'] !=0) & (newdt['Cancelled'] ==1))] #aucune erreur au niveau des vols annulés
```

```python
newdt
```

```python
newdt.describe()
```

```python
#What are the average (mean) departure and arrival delays?

meanDeparture = newdt.describe().iloc[1, 7]
meanArrival = newdt.describe().iloc[1, 10]

print(f"Mean of Departure delay: {meanDeparture} minutes")
print(f"Mean of Arrival delay: {meanArrival} minutes")
```

```python
#Is there a noticable difference in arrival delays for different days of the week?

test = newdt.groupby(['DayOfWeek']).describe()
test.loc[:7, ['DepDelay', 'ArrDelay']]

#As we can see, wednesday and thursday have the highest mean of delay but on saturday, 5 times less of delay than thursday
```

```python
#Which departure airport has the highest average departure delay?

departureDelay =newdt.groupby(['OriginAirportName']).describe()
#departureDelay.sort_values(by='DepDelay')
departureDelay.iloc[:70, 56:64] 
```

```python
#Do late departures tend to result in longer arrival delays than on-time departures?

ontimeDeparture = newdt.loc[newdt['DepDelay'] <= 0] #leave early and on time
ontime = ontimeDeparture.describe().iloc[1, 10]

lateDeparture = newdt.loc[newdt['DepDelay'] >0 ] #leave lately from 1 minute
late = lateDeparture.describe().iloc[1, 10]

print(f"On time(mean): {ontime} minutes   Late(mean): {late} minutes")

#So when the departure is late, are most probably in late during the landing
```

```python
#Which route (from origin airport to destination airport) has the most late* arrivals?*

arrivalDelay =newdt.groupby(['OriginAirportName']).sum()
arrivalDelay.iloc[1,:10]
#departureDelay.sort_values(by='DepDelay')
#arrivalDelay.iloc[:70, 80:88] 
```
